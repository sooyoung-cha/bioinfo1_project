```python
import pandas as pd
cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#', index_col=0)
cnts.head()
```

```python
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
```

```python
# 여기에 코드를 새로 작성해서 논문의 그림과 비슷하게 만들어 봅니다.
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

mask_valid = (cnts['clip_enrichment'] > 0) & (cnts['rden_change'] > 0)
x = np.log2(cnts.loc[mask_valid, 'clip_enrichment'])
y = np.log2(cnts.loc[mask_valid, 'rden_change'])

valid = ~(np.isinf(x) | np.isinf(y))
x = x[valid]
y = y[valid]

r, _ = pearsonr(x, y)

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(x, y, s=1, alpha=0.3, color='black')

ax.set_xlabel('LIN28A CLIP enrichment (log₂)')
ax.set_ylabel('Ribosome density change\nupon Lin28a knockdown (log₂)')
ax.set_xlim(-6, 7)
ax.set_ylim(-4, 3)

ax.text(0.95, 0.05, f'r = {r:.4f}', transform=ax.transAxes,
        ha='right', va='bottom', fontsize=10, style='italic')

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np

cnts['clip_enrichment_log2'] = np.log2(cnts['clip_enrichment'])
cnts['rden_change_log2'] = np.log2(cnts['rden_change'])
cnt_2 = cnts.replace([np.inf, -np.inf], np.nan).dropna(subset=['clip_enrichment_log2', 'rden_change_log2'])
cnts = cnt_2
cnts['rank'] = cnts['clip_enrichment'].rank(ascending=False, method='first')
thrs = [0.05, 0.20, 0.50, 1.00]
thrslist = ['Top 5%', '5-20%', '20-50%', '50-100%']
qs = [cnts['rank'].quantile(q) for q in thrs]
cnts['group'] = pd.cut(cnts['rank'], bins=[0] + qs, labels=thrslist)
cnts.head()
```
